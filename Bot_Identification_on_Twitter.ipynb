```
tweetzipfiles = glob.glob('/content/drive/My Drive/Colab Notebooks/Twitter analysis/*.zip')
tweetzipfiles
```

```
users = []

for tweetzipfile in tweetzipfiles:
  zf = zipfile.ZipFile(tweetzipfile)
  for i, obj in enumerate(zf.infolist()):
    tweets_data = json.load(zf.open(obj))
    if (tweets_data['user']['description']) is not None:
      users.append(tweets_data['user'])

tweets = pd.DataFrame(users)
tweets.head()
```

```
usertweets = tweets.groupby('screen_name')
top_users = usertweets.count()['description'].sort_values(ascending = False)[:1000]
top_users_dict = top_users.to_dict()
user_ordered_dict =sorted(top_users_dict.items(), key=lambda x:x[1])
user_ordered_dict = user_ordered_dict[::-1]
dict_values = []
dict_keys = []
for item in user_ordered_dict[0:1000]:
    dict_keys.append(item[0])
    dict_values.append(item[1])
dict_keys
```

Botometer score of 0.43 as the threshold is being set. This value has been decided after reading the 'Q&A: How Pew Research Center identified bots on Twitter' article by JOHN GRAMLICH which states that Mr.Stefan Wojcik, Pew Research Center computational social scientist has used this value as using a Botometer score that was too high would have meant incorrectly classifying many bots as human accounts – otherwise known as a false negative. On the other hand, if we had set a threshold that was too low, we would have incorrectly labeled lots of human accounts as bots – a false positive.


```
newDict = dict()
botnamesdict = dict()
for (key, value) in bot_dict.items():
  if value != 'None':
    newDict[key] = value
for (key, value) in newDict.items():
  if value > 43:
    botnamesdict[key] = value

print('The potential list of bot names with the corresponding score percentage are:',botnamesdict )
```
