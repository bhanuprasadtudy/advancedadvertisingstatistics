```
from google.colab import drive
drive.mount('/content/drive')
urls=[]
with open('drive/My Drive/Colab Notebooks/salesforce.2018.csv') as input_file:
    for i, line in enumerate(input_file):
        urls.append(line)
urls
```

```
import requests
from newspaper import fulltext
from gensim.summarization import keywords
URL= []
keywordsList=[]
for x in urls[0:9]:
  r = requests.get(x)
  if requests.get(x).status_code == 200:
    try: 
      r = requests.get(x)
      URL.append(x)
      html = r.text
      text = fulltext(html)
      article = keywords(text, lemmatize = True)
      keywordsList.append(article)
    except:
        continue

```

```
import nltk
from nltk.probability import FreqDist


words = nltk.tokenize.word_tokenize(text)
fdist = FreqDist(words)
sorted_d = sorted(fdist.items(), key=lambda x: x[1],reverse=True)
sorted_d
```
